```python
import pandas as pd

orders = pd.read_csv("../data/orders.csv")
products = pd.read_csv("../data/products.csv")
prior = pd.read_csv("../data/order_products__prior.csv")

print("Orders:", orders.shape)
print("Products:", products.shape)
print("Prior:", prior.shape)

orders.head()
prior.head()
products.head()

merged = prior.merge(
    orders[['order_id', 'user_id']],
    on='order_id'
)
print("Merged:", merged.shape)
merged.head()

user_history = merged.groupby(
    'user_id'
)['product_id'].apply(list).to_dict()

print(user_history[1][:10])
print(len(user_history))
```

```python
print(user_history[1][:10])
```

```python
small_users = list(user_history.keys())[:100]
small_user_history = {
    k: user_history[k]
    for k in small_users
}
print(small_user_history)
```

```python
top_items = (merged['product_id']
    .value_counts()
    .head(100)
    .index 
    .tolist())
```

```python
print(top_items)
```

```python
filtered_history = {}

for user, items in small_user_history.items():

    filtered_items = [items for items in items if items in top_items]

    if len(filtered_items) >= 5:
        filtered_history[user] = filtered_items

print(filtered_history)
```

```python
print(len(filtered_history))
```

```python
sample_user = list(filtered_history.keys())[0]

print(filtered_history[sample_user])
```

```python
item_to_index = {
    item: idx
    for idx, item in enumerate(top_items)
}

print(item_to_index)
```

```python
indexed_history = {}

for user, items in filtered_history.items():

    indexed_history[user] = [
        item_to_index[item]
        for item in items
    ]

print(indexed_history[sample_user])
```

```python
import pickle

with open(
    "../data/processed/indexed_history.pkl",
    "wb"
) as f:

    pickle.dump(indexed_history, f)
```
